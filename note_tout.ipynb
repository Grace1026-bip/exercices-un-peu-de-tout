```python
# pour les fonction fonctionnel on utilise def
# les tructures conditionnelles
# masse, hauteur et g sont les q-word soit le prete nom des n
def e_potentielle (masse, e, hauteur, g=9.81):
	E= masse * hauteur*g
	print("l'énergie potentielle est ", E)
	# return permet aux fonction de produire un résultat
	if E>e :
		return E
	elif E==0:
		return 0
	else:
		return e
	
resultat = e_potentielle(5, 10, 1)
print(resultat)
```

```python
#la structure conditionnelle(if,elif,else)
def signe(x):
	if x>0:
		print(x, "est posif")
	elif x==0:
		print(x, "est nul")
	else:
		print(x, "est négatif")

signe(5)
```

```python
#les boucles

for i in range(10, -1):
	signe(i)

#while a <10:
	#print(a)
	#a += 1

def fibonacci(n):
	for i in range (0, n+1):
		signe(i)

fibonacci(10)

def decompte(n):
	a=0
	b=1
	while a <n:
		a, b = b, a+b
decompte(10)
```

```python
#nested list(liste dans une liste)
#liste compréhension
#liste normal
liste_com =[]
for i in range(10):
	liste_com.append(i**2)
print(liste_com)

#liste en compréhension
liste_com1=[i**2 for i in range(10)] #liste en compréhension plus rapide et plus professionnelle
print(liste_com1)

#liste_com2 = [[i for range(5)] for j in range(4)]
#print(liste_com2)

#dictionnaire comprehension = c'est qui est inclus(do this)+boucle(for)+conditions
nom =['Pierre','Jean','Jules']
age=[24, 39,34]

dico={g:n for g,n in enumerate(nom) }
print(dico)

dico2={nom:age for nom, age in zip(nom,age) if age > 30}
print(dico2)

tuple2 = tuple((i**2 for i in range (5)))
print(tuple2)

#tp
h=[i for i in range(21)]
p =[j**2 for j in h]
dico3={h:p for h, p in zip(h,p)}
print(dico3)
#resolution
carres_pairs={str(k) : k**2 for k in range(0,20)}
print(carres_pairs)
```
